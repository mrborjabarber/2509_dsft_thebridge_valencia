# ðŸ“ 1. Vectores: Los Bloques de ConstrucciÃ³n

## Â¿QuÃ© es un vector?

Un vector es una **lista ordenada de nÃºmeros** que representa informaciÃ³n en el espacio multidimensional.

### Ejemplos del mundo real:
- ðŸ‘¤ **Perfil de cliente**: `[edad, ingresos, gastos]` = `[25, 50000, 30000]`
- ðŸ  **CaracterÃ­sticas de casa**: `[mÂ², habitaciones, baÃ±os]` = `[120, 3, 2]`
- ðŸ–¼ï¸ **PÃ­xel de imagen**: `[rojo, verde, azul]` = `[255, 128, 64]`

---

## ðŸ—ï¸ CreaciÃ³n de Vectores

```python
# === CREAR VECTORES REPRESENTANDO CLIENTES BANCARIOS ===
# Formato: [edad, ingresos_anuales, gastos_anuales]

cliente_a = np.array([28, 55000, 35000])  # Cliente joven profesional
cliente_b = np.array([35, 70000, 45000])  # Cliente establecido
cliente_c = np.array([42, 85000, 52000])  # Cliente senior

print("ðŸ¦ === PERFILES DE CLIENTES BANCARIOS ===")
print(f"Cliente A: {cliente_a} â†’ {cliente_a[0]} aÃ±os, ${cliente_a[1]:,} ingresos")
print(f"Cliente B: {cliente_b} â†’ {cliente_b[0]} aÃ±os, ${cliente_b[1]:,} ingresos")
print(f"Cliente C: {cliente_c} â†’ {cliente_c[0]} aÃ±os, ${cliente_c[1]:,} ingresos")

# === PROPIEDADES BÃSICAS ===
print(f"\nðŸ“Š === PROPIEDADES DEL VECTOR ===")
print(f"Forma del vector: {cliente_a.shape}")
print(f"NÃºmero de dimensiones: {cliente_a.ndim}")
print(f"Tipo de datos: {cliente_a.dtype}")
print(f"NÃºmero de elementos: {cliente_a.size}")

# === DIFERENTES FORMAS DE CREAR VECTORES ===
print(f"\nðŸ› ï¸ === FORMAS DE CREAR VECTORES ===")

vector_ceros = np.zeros(4)  # Vector de ceros
vector_unos = np.ones(3)    # Vector de unos
vector_rango = np.arange(5, 10)  # Vector secuencial

np.random.seed(42)  # Para reproducibilidad
vector_aleatorio = np.random.normal(100, 15, 5)  # Media=100, std=15

print(f"Vector de ceros: {vector_ceros}")
print(f"Vector de unos: {vector_unos}")
print(f"Vector rango (5-9): {vector_rango}")
print(f"Vector aleatorio: {vector_aleatorio.round(1)}")
```

## âž• Operaciones BÃ¡sicas con Vectores

```python
# === SUMA Y RESTA DE VECTORES ===
print("âž• === SUMA DE VECTORES ===")

# La suma combina caracterÃ­sticas de ambos clientes
suma_clientes = cliente_a + cliente_b
promedio_clientes = suma_clientes / 2

print(f"Cliente A: {cliente_a}")
print(f"Cliente B: {cliente_b}")
print(f"Suma:      {suma_clientes}")
print(f"Promedio:  {promedio_clientes.astype(int)}")

print(f"\nðŸ’¡ Cliente promedio: {promedio_clientes[0]:.0f} aÃ±os, ${promedio_clientes[1]:,.0f} ingresos")

# === RESTA DE VECTORES ===
print(f"\nâž– === DIFERENCIAS ENTRE CLIENTES ===")
diferencia = cliente_b - cliente_a
print(f"Diferencia B - A: {diferencia}")
print(f"ðŸ’° Cliente B gana ${diferencia[1]:,} mÃ¡s que Cliente A")
print(f"ðŸŽ‚ Cliente B es {diferencia[0]} aÃ±os mayor que Cliente A")

# === MULTIPLICACIÃ“N POR ESCALAR ===
print(f"\nâœ–ï¸ === MULTIPLICACIÃ“N POR ESCALAR ===")
# Simular inflaciÃ³n del 5%
cliente_a_inflacion = cliente_a * 1.05
print(f"Cliente A original: {cliente_a}")
print(f"Con inflaciÃ³n 5%:   {cliente_a_inflacion.astype(int)}")

# ConversiÃ³n de moneda (USD a EUR, aprox 0.85)
cliente_a_euros = cliente_a.copy().astype(float)
cliente_a_euros[1:] *= 0.85  # Solo convertir valores monetarios
print(f"En euros:           {cliente_a_euros.astype(int)}")
```

## ðŸ“ Magnitud y NormalizaciÃ³n

```python
# === MAGNITUD (NORMA) DEL VECTOR ===
print("ðŸ“ === MAGNITUD DE VECTORES ===")

# La magnitud mide quÃ© tan "grande" es el vector
magnitud_a = np.linalg.norm(cliente_a)
magnitud_b = np.linalg.norm(cliente_b) 
magnitud_c = np.linalg.norm(cliente_c)

print(f"Magnitud Cliente A: {magnitud_a:,.0f}")
print(f"Magnitud Cliente B: {magnitud_b:,.0f}")
print(f"Magnitud Cliente C: {magnitud_c:,.0f}")

# CÃ¡lculo manual de la magnitud para entender mejor
magnitud_manual = np.sqrt(cliente_a[0]**2 + cliente_a[1]**2 + cliente_a[2]**2)
print(f"\nðŸ”¢ CÃ¡lculo manual Cliente A:")
print(f"âˆš({cliente_a[0]}Â² + {cliente_a[1]}Â² + {cliente_a[2]}Â²) = {magnitud_manual:,.0f}")

# === DIFERENTES TIPOS DE NORMAS ===
print(f"\nðŸ“ === TIPOS DE NORMAS ===")

norma_l1 = np.linalg.norm(cliente_a, ord=1)      # Suma de valores absolutos
norma_l2 = np.linalg.norm(cliente_a, ord=2)      # Euclidiana (default)
norma_inf = np.linalg.norm(cliente_a, ord=np.inf) # Valor mÃ¡ximo

print(f"Norma L1 (Manhattan):  {norma_l1:,.0f}")
print(f"Norma L2 (Euclidiana): {norma_l2:,.0f}")
print(f"Norma Lâˆž (MÃ¡xima):     {norma_inf:,.0f}")

# === NORMALIZACIÃ“N ===
print(f"\nðŸŽ¯ === NORMALIZACIÃ“N DE VECTORES ===")

# Normalizar = crear vector unitario (magnitud = 1)
cliente_a_normalizado = cliente_a / np.linalg.norm(cliente_a)
cliente_b_normalizado = cliente_b / np.linalg.norm(cliente_b)

print(f"Cliente A normalizado: [{cliente_a_normalizado[0]:.3f}, {cliente_a_normalizado[1]:.3f}, {cliente_a_normalizado[2]:.3f}]")
print(f"Nueva magnitud: {np.linalg.norm(cliente_a_normalizado):.10f}")
print(f"ðŸ’¡ Los vectores normalizados siempre tienen magnitud = 1")
```

## ðŸŽ¯ Producto Punto: La OperaciÃ³n Estrella

```python
# === PRODUCTO PUNTO ===
print("ðŸŽ¯ === PRODUCTO PUNTO ===")

# El producto punto mide quÃ© tan "similares" son dos vectores
producto_ab = np.dot(cliente_a, cliente_b)
producto_ac = np.dot(cliente_a, cliente_c)
producto_bc = np.dot(cliente_b, cliente_c)

print(f"Cliente A Â· Cliente B = {producto_ab:,}")
print(f"Cliente A Â· Cliente C = {producto_ac:,}")
print(f"Cliente B Â· Cliente C = {producto_bc:,}")

# CÃ¡lculo manual para entender
producto_manual = (cliente_a[0] * cliente_b[0] + 
                  cliente_a[1] * cliente_b[1] + 
                  cliente_a[2] * cliente_b[2])

print(f"\nðŸ”¢ CÃ¡lculo manual AÂ·B:")
print(f"{cliente_a[0]}Ã—{cliente_b[0]} + {cliente_a[1]}Ã—{cliente_b[1]} + {cliente_a[2]}Ã—{cliente_b[2]} = {producto_manual:,}")

# === SIMILITUD COSENO ===
print(f"\nðŸ”„ === SIMILITUD COSENO ===")

# Similitud coseno = producto punto de vectores normalizados
sim_ab = np.dot(cliente_a_normalizado, cliente_b_normalizado)
sim_ac = np.dot(cliente_a / np.linalg.norm(cliente_a), cliente_c / np.linalg.norm(cliente_c))

print(f"Similitud A-B: {sim_ab:.3f}")
print(f"Similitud A-C: {sim_ac:.3f}")

# Convertir a Ã¡ngulo para mejor interpretaciÃ³n
angulo_ab = np.degrees(np.arccos(np.clip(sim_ab, -1, 1)))
angulo_ac = np.degrees(np.arccos(np.clip(sim_ac, -1, 1)))

print(f"\nðŸ“ Ãngulos entre vectores:")
print(f"Cliente A-B: {angulo_ab:.1f}Â°")
print(f"Cliente A-C: {angulo_ac:.1f}Â°")

# === INTERPRETACIÃ“N PARA NEGOCIO ===
print(f"\nðŸ’¼ === APLICACIÃ“N EN MARKETING ===")

def interpretar_similitud(similitud, nombre1, nombre2):
    if similitud > 0.9:
        return f"âœ… {nombre1} y {nombre2} son muy similares - usar misma estrategia"
    elif similitud > 0.7:
        return f"ðŸŸ¡ {nombre1} y {nombre2} son similares - estrategias relacionadas"
    elif similitud > 0.5:
        return f"ðŸŸ  {nombre1} y {nombre2} son algo similares - enfoques diferentes"
    else:
        return f"ðŸ”´ {nombre1} y {nombre2} son muy diferentes - estrategias distintas"

print(interpretar_similitud(sim_ab, "Cliente A", "Cliente B"))
print(interpretar_similitud(sim_ac, "Cliente A", "Cliente C"))
```

## ðŸ“Š VisualizaciÃ³n de Vectores 2D

```python
# === VISUALIZACIÃ“N DE VECTORES ===
print("ðŸ“Š === VISUALIZACIÃ“N DE VECTORES 2D ===")

# Crear vectores 2D mÃ¡s simples para visualizaciÃ³n
v1 = np.array([3, 4])    # Vector ejemplo 1
v2 = np.array([1, 2])    # Vector ejemplo 2
v3 = v1 + v2             # Suma vectorial

print(f"Vector 1: {v1}")
print(f"Vector 2: {v2}")
print(f"Suma:     {v3}")

# Crear visualizaciÃ³n
plt.figure(figsize=(10, 8))

# Dibujar vectores desde el origen
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.008, label=f'Vector 1: {v1}', alpha=0.8)

plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, 
           color='blue', width=0.008, label=f'Vector 2: {v2}', alpha=0.8)

plt.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, 
           color='green', width=0.01, label=f'Suma: {v3}', alpha=0.9)

# Dibujar la regla del paralelogramo
plt.quiver(v1[0], v1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, 
           color='blue', alpha=0.4, width=0.005, linestyle='--')

plt.quiver(v2[0], v2[1], v1[0], v1[1], angles='xy', scale_units='xy', scale=1, 
           color='red', alpha=0.4, width=0.005, linestyle='--')

# ConfiguraciÃ³n del grÃ¡fico
plt.xlim(-0.5, 6)
plt.ylim(-0.5, 7)
plt.grid(True, alpha=0.3)
plt.legend(loc='upper left')
plt.title('ðŸ” Suma de Vectores: Regla del Paralelogramo', fontsize=14, fontweight='bold')
plt.xlabel('Componente X')
plt.ylabel('Componente Y')

# Marcar puntos importantes
plt.plot(0, 0, 'ko', markersize=8, label='Origen')
plt.plot(v1[0], v1[1], 'ro', markersize=8, alpha=0.7)
plt.plot(v2[0], v2[1], 'bo', markersize=8, alpha=0.7)
plt.plot(v3[0], v3[1], 'go', markersize=10, alpha=0.9)

plt.tight_layout()
plt.show()

# === VERIFICACIÃ“N DE PROPIEDADES ===
print(f"\nâœ… === PROPIEDADES VECTORIALES ===")
print(f"Conmutatividad: v1 + v2 = {v1 + v2}, v2 + v1 = {v2 + v1}")
print(f"Â¿Son iguales? {np.array_equal(v1 + v2, v2 + v1)}")

producto_2d = np.dot(v1, v2)
angulo_2d = np.degrees(np.arccos(producto_2d / (np.linalg.norm(v1) * np.linalg.norm(v2))))
print(f"\nProducto punto: {producto_2d}")
print(f"Ãngulo entre vectores: {angulo_2d:.1f}Â°")
```

# ðŸ“Š 2. Matrices: Organizando Datos Multidimensionales

## Â¿QuÃ© es una matriz?

Una matriz es una **tabla rectangular de nÃºmeros** que organiza mÃºltiples vectores de forma estructurada.

### En data science:
- **Filas** = Observaciones, muestras, ejemplos
- **Columnas** = CaracterÃ­sticas, variables, features

### Ejemplos:
- ðŸ¦ **Dataset bancario**: [clientes Ã— caracterÃ­sticas_financieras]
- ðŸ–¼ï¸ **Imagen**: [pÃ­xeles_alto Ã— pÃ­xeles_ancho]
- ðŸ“º **Netflix**: [usuarios Ã— pelÃ­culas_valoradas]
- ðŸ›’ **E-commerce**: [productos Ã— caracterÃ­sticas]

---

## ðŸ—ï¸ CreaciÃ³n y Propiedades de Matrices

```python
# === CREAR MATRIZ DE DATOS BANCARIOS ===
print("ðŸ¦ === DATASET BANCARIO COMO MATRIZ ===")

# Cada fila = un cliente, cada columna = una caracterÃ­stica
# Columnas: [edad, ingresos_anuales, gastos_anuales, score_crediticio, aÃ±os_cliente]
clientes_matriz = np.array([
    [25, 50000, 30000, 720, 2],   # Cliente joven profesional
    [35, 70000, 45000, 780, 5],   # Cliente establecido
    [28, 55000, 33000, 690, 3],   # Cliente promedio
    [42, 85000, 52000, 820, 8],   # Cliente premium
    [30, 60000, 38000, 750, 4],   # Cliente en crecimiento
    [38, 75000, 48000, 800, 6]    # Cliente consolidado
])

print("Matriz de clientes bancarios:")
print(clientes_matriz)

# === PROPIEDADES DE LA MATRIZ ===
filas, columnas = clientes_matriz.shape
print(f"\nðŸ“Š === PROPIEDADES DE LA MATRIZ ===")
print(f"Forma (shape): {clientes_matriz.shape}")
print(f"NÃºmero de clientes (filas): {filas}")
print(f"NÃºmero de caracterÃ­sticas (columnas): {columnas}")
print(f"Total de elementos: {clientes_matriz.size}")
print(f"Dimensiones: {clientes_matriz.ndim}D")
print(f"Tipo de datos: {clientes_matriz.dtype}")
print(f"Memoria utilizada: {clientes_matriz.nbytes} bytes")

# === CREAR DATAFRAME PARA MEJOR VISUALIZACIÃ“N ===
caracteristicas = ['Edad', 'Ingresos', 'Gastos', 'Score_Credit', 'AÃ±os_Cliente']
indices_clientes = [f'Cliente_{i+1}' for i in range(filas)]

df_clientes = pd.DataFrame(clientes_matriz, 
                          columns=caracteristicas,
                          index=indices_clientes)

print(f"\nðŸ“‹ === VISUALIZACIÃ“N COMO DATAFRAME ===")
print(df_clientes)

# === ACCESO A DATOS ===
print(f"\nðŸ” === ACCESO A ELEMENTOS ===")
print(f"Primer cliente (fila 0): {clientes_matriz[0]}")
print(f"Todos los ingresos (columna 1): {clientes_matriz[:, 1]}")
print(f"Score del Cliente 3: {clientes_matriz[2, 3]}")
print(f"Primeros 3 clientes:\n{clientes_matriz[:3]}")
print(f"Solo datos financieros (columnas 1-2):\n{clientes_matriz[:, 1:3]}")
```

## ðŸ“ˆ AnÃ¡lisis EstadÃ­stico de la Matriz

```python
# === ANÃLISIS ESTADÃSTICO ===
print("ðŸ“ˆ === ANÃLISIS ESTADÃSTICO POR CARACTERÃSTICA ===")

for i, caracteristica in enumerate(caracteristicas):
    valores = clientes_matriz[:, i]
    print(f"\n{caracteristica}:")
    print(f"  â€¢ Promedio: {np.mean(valores):,.1f}")
    print(f"  â€¢ Mediana: {np.median(valores):,.1f}")
    print(f"  â€¢ MÃ­nimo: {np.min(valores):,.0f}")
    print(f"  â€¢ MÃ¡ximo: {np.max(valores):,.0f}")
    print(f"  â€¢ DesviaciÃ³n estÃ¡ndar: {np.std(valores):,.1f}")
    print(f"  â€¢ Rango: {np.max(valores) - np.min(valores):,.0f}")

# === ESTADÃSTICAS GLOBALES ===
print(f"\nðŸŒ === ESTADÃSTICAS GLOBALES ===")
print(f"Suma total: {np.sum(clientes_matriz):,.0f}")
print(f"Promedio global: {np.mean(clientes_matriz):,.1f}")
print(f"Valor mÃ¡ximo: {np.max(clientes_matriz):,.0f}")
print(f"Valor mÃ­nimo: {np.min(clientes_matriz):,.0f}")

# === ENCONTRAR POSICIONES EXTREMAS ===
pos_max = np.unravel_index(np.argmax(clientes_matriz), clientes_matriz.shape)
pos_min = np.unravel_index(np.argmin(clientes_matriz), clientes_matriz.shape)

print(f"\nðŸŽ¯ === VALORES EXTREMOS ===")
print(f"Valor mÃ¡ximo {np.max(clientes_matriz)} en: {indices_clientes[pos_max[0]]}, {caracteristicas[pos_max[1]]}")
print(f"Valor mÃ­nimo {np.min(clientes_matriz)} en: {indices_clientes[pos_min[0]]}, {caracteristicas[pos_min[1]]}")

# === CORRELACIONES SIMPLES ===
print(f"\nðŸ”— === ANÃLISIS DE CORRELACIÃ“N ===")
# CorrelaciÃ³n entre ingresos y gastos
ingresos = clientes_matriz[:, 1]
gastos = clientes_matriz[:, 2]
correlacion = np.corrcoef(ingresos, gastos)[0, 1]

print(f"CorrelaciÃ³n Ingresos-Gastos: {correlacion:.3f}")
if correlacion > 0.7:
    print("ðŸ’¡ CorrelaciÃ³n fuerte: A mayor ingreso, mayor gasto")
elif correlacion > 0.3:
    print("ðŸ’¡ CorrelaciÃ³n moderada: RelaciÃ³n parcial entre ingreso y gasto")
else:
    print("ðŸ’¡ CorrelaciÃ³n dÃ©bil: Poca relaciÃ³n entre ingreso y gasto")
```

## âš™ï¸ Operaciones BÃ¡sicas con Matrices

```python
# === TRANSPUESTA DE MATRIZ ===
print("ðŸ”„ === TRANSPUESTA DE MATRIZ ===")

clientes_T = clientes_matriz.T
print(f"Matriz original: {clientes_matriz.shape} (clientes Ã— caracterÃ­sticas)")
print(f"Matriz transpuesta: {clientes_T.shape} (caracterÃ­sticas Ã— clientes)")
print(f"\nPrimeras 3 columnas de la transpuesta:")
print(clientes_T[:, :3])

# === OPERACIONES ELEMENT-WISE ===
print(f"\nâž• === OPERACIONES ELEMENT-WISE ===")

# Simular datos del aÃ±o anterior (5% menos)
clientes_aÃ±o_anterior = clientes_matriz * 0.95

# Suma de matrices
suma_matrices = clientes_matriz + clientes_aÃ±o_anterior
print(f"Suma de matrices (actual + anterior):")
print(f"Cliente 1 actual: {clientes_matriz[0]}")
print(f"Cliente 1 anterior: {clientes_aÃ±o_anterior[0].astype(int)}")
print(f"Cliente 1 suma: {suma_matrices[0].astype(int)}")

# Diferencia y crecimiento
diferencia_anual = clientes_matriz - clientes_aÃ±o_anterior
crecimiento_porcentual = (diferencia_anual / clientes_aÃ±o_anterior) * 100

print(f"\nCrecimiento anual del Cliente 1:")
for i, caracteristica in enumerate(caracteristicas):
    print(f"  â€¢ {caracteristica}: {crecimiento_porcentual[0, i]:.1f}%")

# === MULTIPLICACIÃ“N POR ESCALAR ===
print(f"\nâœ–ï¸ === MULTIPLICACIÃ“N POR ESCALAR ===")

# Ajuste por inflaciÃ³n
factor_inflacion = 1.03
clientes_ajustados = clientes_matriz * factor_inflacion

print(f"Ajuste por inflaciÃ³n del 3%:")
print(f"Cliente 1 original: {clientes_matriz[0]}")
print(f"Cliente 1 ajustado: {clientes_ajustados[0].astype(int)}")

# ConversiÃ³n selectiva (solo columnas monetarias)
tipo_cambio = 0.85
clientes_euros = clientes_matriz.copy().astype(float)
clientes_euros[:, 1:3] *= tipo_cambio  # Solo ingresos y gastos

print(f"\nConversiÃ³n a euros (solo valores monetarios):")
print(f"Cliente 1 USD: {clientes_matriz[0]}")
print(f"Cliente 1 EUR: {clientes_euros[0].astype(int)}")
```

## ðŸ” Indexing y Filtrado Avanzado

## âš¡ MultiplicaciÃ³n de Matrices: El CorazÃ³n del ML

```python
# === MULTIPLICACIÃ“N DE MATRICES ===
print("âš¡ === MULTIPLICACIÃ“N DE MATRICES ===")

# Crear sistema de scoring crediticio
# Cada fila = un modelo diferente, cada columna = peso de una caracterÃ­stica
modelos_scoring = np.array([
    [0.1,  0.3, -0.2,  0.4,  0.1],   # Modelo conservador
    [0.05, 0.4, -0.3,  0.3,  0.15],  # Modelo orientado a ingresos
    [0.2,  0.2, -0.1,  0.35, 0.25]   # Modelo que valora experiencia
])

print("ðŸ¦ Modelos de scoring crediticio (pesos):")
modelos_df = pd.DataFrame(modelos_scoring, 
                         columns=caracteristicas,
                         index=['Conservador', 'Ingresos', 'Experiencia'])
print(modelos_df)

# Aplicar modelos: Clientes Ã— Modelos^T = Puntuaciones
puntuaciones = np.dot(clientes_matriz, modelos_scoring.T)

print(f"\nðŸ“Š === RESULTADOS DEL SCORING ===")
print(f"Forma clientes: {clientes_matriz.shape}")
print(f"Forma modelos^T: {modelos_scoring.T.shape}")
print(f"Forma resultado: {puntuaciones.shape}")

# Crear DataFrame de puntuaciones
puntuaciones_df = pd.DataFrame(puntuaciones,
                              columns=['Conservador', 'Ingresos', 'Experiencia'],
                              index=indices_clientes)

print("\nPuntuaciones por cliente y modelo:")
print(puntuaciones_df.round(0))

# === ANÃLISIS DE RESULTADOS ===
print(f"\nðŸ† === ANÃLISIS DE MEJORES CLIENTES ===")
mejor_por_modelo = puntuaciones_df.idxmax()
puntuacion_maxima = puntuaciones_df.max()

for modelo in puntuaciones_df.columns:
    mejor_cliente = mejor_por_modelo[modelo]
    score = puntuacion_maxima[modelo]
    print(f"Modelo {modelo}: {mejor_cliente} (score: {score:.0f})")

# === EJEMPLO DE CÃLCULO MANUAL ===
print(f"\nðŸ”¢ === VERIFICACIÃ“N MANUAL ===")
cliente_1 = clientes_matriz[0]
modelo_conservador = modelos_scoring[0]

calculo_manual = np.sum(cliente_1 * modelo_conservador)
calculo_numpy = puntuaciones[0, 0]

print(f"Cliente 1: {cliente_1}")
print(f"Modelo conservador: {modelo_conservador}")
print(f"CÃ¡lculo manual: {calculo_manual:.0f}")
print(f"Resultado NumPy: {calculo_numpy:.0f}")
print(f"Â¿Son iguales? {np.isclose(calculo_manual, calculo_numpy)}")
```

## ðŸ”„ Matriz Identidad, Inversa y Determinante

```python
# === CONCEPTOS AVANZADOS DE MATRICES ===
print("ðŸ”„ === MATRIZ IDENTIDAD ===")

# La matriz identidad es como el "1" para las matrices
I_2 = np.eye(2)  # Identidad 2x2
I_3 = np.eye(3)  # Identidad 3x3

print(f"Matriz identidad 2Ã—2:\n{I_2}")
print(f"\nMatriz identidad 3Ã—3:\n{I_3}")

# Demostrar propiedad: A Ã— I = A
A_ejemplo = np.array([[2, 3], [4, 5]])
resultado = np.dot(A_ejemplo, I_2)

print(f"\nâœ… Propiedad A Ã— I = A:")
print(f"A:\n{A_ejemplo}")
print(f"A Ã— I:\n{resultado}")
print(f"Â¿Son iguales? {np.array_equal(A_ejemplo, resultado)}")

# === DETERMINANTE ===
print(f"\nðŸŽ¯ === DETERMINANTE ===")

matrices_ejemplo = {
    "Invertible": np.array([[2, 1], [1, 1]]),
    "No invertible": np.array([[2, 4], [1, 2]]),
    "Identidad": I_2,
    "Escalar 3x": 3 * I_2
}

for nombre, matriz in matrices_ejemplo.items():
    det = np.linalg.det(matriz)
    print(f"\n{nombre}:")
    print(f"Matriz:\n{matriz}")
    print(f"Determinante: {det:.3f}")
    
    if abs(det) > 1e-10:
        print(f"  âœ… Invertible (det â‰  0)")
        print(f"  ðŸ“ Factor de escalamiento: {abs(det):.1f}x")
    else:
        print(f"  âŒ No invertible (det = 0)")

# === MATRIZ INVERSA ===
print(f"\nðŸ”„ === MATRIZ INVERSA ===")

A = matrices_ejemplo["Invertible"]
if abs(np.linalg.det(A)) > 1e-10:
    A_inv = np.linalg.inv(A)
    
    print(f"Matriz A:\n{A}")
    print(f"Matriz Aâ»Â¹:\n{A_inv}")
    
    # Verificar A Ã— Aâ»Â¹ = I
    producto = np.dot(A, A_inv)
    print(f"\nA Ã— Aâ»Â¹ (debe ser identidad):\n{np.round(producto, 10)}")
    print(f"Â¿Es identidad? {np.allclose(producto, I_2)}")

# === APLICACIÃ“N: RESOLVER SISTEMA DE ECUACIONES ===
print(f"\nðŸ§® === RESOLVER SISTEMA DE ECUACIONES ===")

# Sistema: 2x + y = 5, x + y = 3
A_sistema = np.array([[2, 1], [1, 1]])
b_sistema = np.array([5, 3])

print(f"Sistema de ecuaciones:")
print(f"2x + y = 5")
print(f"x + y = 3")
print(f"\nEn forma matricial: A Ã— x = b")
print(f"A =\n{A_sistema}")
print(f"b = {b_sistema}")

# Resolver usando la inversa
x_solucion = np.linalg.solve(A_sistema, b_sistema)
print(f"\nâœ… SoluciÃ³n: x = {x_solucion}")
print(f"ðŸ’¡ x = {x_solucion[0]:.1f}, y = {x_solucion[1]:.1f}")

# Verificar la soluciÃ³n
verificacion = np.dot(A_sistema, x_solucion)
print(f"VerificaciÃ³n A Ã— x = {verificacion}")
print(f"Â¿Correcto? {np.allclose(verificacion, b_sistema)}")
```

# ðŸŽ¯ 3. AplicaciÃ³n PrÃ¡ctica: RegresiÃ³n Lineal

## El Poder del Ãlgebra Matricial

Vamos a ver cÃ³mo el Ã¡lgebra matricial convierte un problema complejo (encontrar la mejor lÃ­nea para un conjunto de datos) en una **operaciÃ³n elegante**.

### La fÃ³rmula mÃ¡gica:
**Î² = (X^T X)â»Â¹ X^T y**

Esta ecuaciÃ³n resuelve de forma exacta el problema de regresiÃ³n lineal sin iteraciones ni aproximaciones.

---

## ðŸ  Ejemplo: PredicciÃ³n de Precios de Casas

```python
# === GENERAR DATOS REALISTAS ===
print("ðŸ  === PREDICCIÃ“N DE PRECIOS DE CASAS ===")

# Generar dataset sintÃ©tico pero realista
np.random.seed(42)
n_casas = 100

# TamaÃ±os de casas entre 50 y 300 mÂ²
tamaÃ±o_casas = np.random.uniform(50, 300, n_casas)

# Precio = 1500â‚¬ por mÂ² + intercepto + ruido
precio_real = 20000 + 1500 * tamaÃ±o_casas + np.random.normal(0, 15000, n_casas)
precio_real = np.maximum(precio_real, 50000)  # Precios mÃ­nimos

print(f"Dataset generado: {n_casas} casas")
print(f"TamaÃ±o promedio: {np.mean(tamaÃ±o_casas):.1f} mÂ²")
print(f"Precio promedio: â‚¬{np.mean(precio_real):,.0f}")
print(f"\nEjemplos:")
for i in range(5):
    print(f"  Casa {i+1}: {tamaÃ±o_casas[i]:.0f} mÂ² â†’ â‚¬{precio_real[i]:,.0f}")

# === PREPARAR DATOS PARA REGRESIÃ“N ===
print(f"\nðŸ”§ === PREPARACIÃ“N DE DATOS ===")

# Matriz X: columna de unos (intercepto) + tamaÃ±os
X = np.column_stack([np.ones(n_casas), tamaÃ±o_casas])
y = precio_real

print(f"Matriz X: {X.shape} (casas Ã— [intercepto, tamaÃ±o])")
print(f"Vector y: {y.shape} (precios)")
print(f"\nPrimeras 5 filas de X:")
print(X[:5])
print(f"Primeros 5 precios: {y[:5].astype(int)}")
```

## ðŸ§® Aplicando la FÃ³rmula Matricial

```python
# === APLICAR LA FÃ“RMULA: Î² = (X^T X)^(-1) X^T y ===
print("ðŸ§® === REGRESIÃ“N CON ÃLGEBRA MATRICIAL ===")

# Paso a paso para entender cada operaciÃ³n
print("Aplicando Î² = (X^T X)^(-1) X^T y paso a paso:")

# Paso 1: X^T
X_T = X.T
print(f"\n1. X^T: {X_T.shape}")

# Paso 2: X^T X
XtX = np.dot(X_T, X)
print(f"2. X^T X: {XtX.shape}")
print(f"   X^T X =\n{XtX}")

# Paso 3: (X^T X)^(-1)
XtX_inv = np.linalg.inv(XtX)
print(f"\n3. (X^T X)^(-1):")
print(f"   {XtX_inv}")

# Paso 4: X^T y
Xty = np.dot(X_T, y)
print(f"\n4. X^T y: {Xty}")

# Paso 5: Î² = (X^T X)^(-1) X^T y
beta = np.dot(XtX_inv, Xty)
intercepto, pendiente = beta

print(f"\n5. Î² = (X^T X)^(-1) X^T y:")
print(f"   Î² = {beta}")
print(f"   ðŸ“Š Intercepto: â‚¬{intercepto:,.0f}")
print(f"   ðŸ“ˆ Pendiente: â‚¬{pendiente:,.0f} por mÂ²")

# === INTERPRETACIÃ“N DEL MODELO ===
print(f"\nðŸ  === MODELO FINAL ===")
print(f"EcuaciÃ³n: Precio = {intercepto:,.0f} + {pendiente:,.0f} Ã— TamaÃ±o")
print(f"")
print(f"ðŸ’¡ Significado:")
print(f"   â€¢ Precio base: â‚¬{intercepto:,.0f}")
print(f"   â€¢ Precio por mÂ²: â‚¬{pendiente:,.0f}")
print(f"   â€¢ Casa de 100 mÂ²: â‚¬{intercepto + pendiente * 100:,.0f}")
print(f"   â€¢ Casa de 200 mÂ²: â‚¬{intercepto + pendiente * 200:,.0f}")
```

## ðŸ“Š EvaluaciÃ³n del Modelo

```python
# === EVALUACIÃ“N DEL MODELO ===
print("ðŸ“Š === EVALUACIÃ“N DEL MODELO ===")

# Hacer predicciones
y_predicho = np.dot(X, beta)

# Calcular mÃ©tricas
errores = y - y_predicho
mse = np.mean(errores**2)  # Error cuadrÃ¡tico medio
rmse = np.sqrt(mse)        # RaÃ­z del error cuadrÃ¡tico medio
mae = np.mean(np.abs(errores))  # Error absoluto medio

# RÂ² (coeficiente de determinaciÃ³n)
ss_res = np.sum(errores**2)  # Suma cuadrados residuales
ss_tot = np.sum((y - np.mean(y))**2)  # Suma cuadrados totales
r2 = 1 - (ss_res / ss_tot)

print(f"MÃ©tricas de rendimiento:")
print(f"  â€¢ MSE: {mse:,.0f}")
print(f"  â€¢ RMSE: â‚¬{rmse:,.0f}")
print(f"  â€¢ MAE: â‚¬{mae:,.0f}")
print(f"  â€¢ RÂ²: {r2:.3f} ({r2:.1%})")

# InterpretaciÃ³n del RÂ²
if r2 > 0.9:
    calidad = "ðŸŸ¢ Excelente"
elif r2 > 0.7:
    calidad = "ðŸŸ¡ Bueno"
elif r2 > 0.5:
    calidad = "ðŸŸ  Aceptable"
else:
    calidad = "ðŸ”´ Pobre"

print(f"\nðŸ’¡ {calidad}: {r2:.1%} de la variaciÃ³n se explica por el tamaÃ±o")

# === COMPARACIÃ“N CON SKLEARN ===
print(f"\nðŸ” === VERIFICACIÃ“N CON SCIKIT-LEARN ===")

modelo_sklearn = LinearRegression()
modelo_sklearn.fit(tamaÃ±o_casas.reshape(-1, 1), y)

print(f"Nuestro cÃ¡lculo:")
print(f"  â€¢ Intercepto: â‚¬{intercepto:,.0f}")
print(f"  â€¢ Pendiente: â‚¬{pendiente:,.0f}")
print(f"  â€¢ RÂ²: {r2:.6f}")

print(f"\nScikit-learn:")
print(f"  â€¢ Intercepto: â‚¬{modelo_sklearn.intercept_:,.0f}")
print(f"  â€¢ Pendiente: â‚¬{modelo_sklearn.coef_[0]:,.0f}")
print(f"  â€¢ RÂ²: {modelo_sklearn.score(tamaÃ±o_casas.reshape(-1, 1), y):.6f}")

diferencia_intercepto = abs(intercepto - modelo_sklearn.intercept_)
diferencia_pendiente = abs(pendiente - modelo_sklearn.coef_[0])

print(f"\nâœ… Diferencias:")
print(f"  â€¢ Intercepto: â‚¬{diferencia_intercepto:.2f}")
print(f"  â€¢ Pendiente: â‚¬{diferencia_pendiente:.2f}")
print(f"\nðŸŽ¯ Â¡Ãlgebra matricial produce resultados idÃ©nticos!")
```
